```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
fontdef = {'family' : 'Arial',
        'weight' : 'normal',
        'size'   : 12}

font = {'family' : 'Arial',
        'weight' : 'bold',
        'size'   : 16}


matplotlib.rc('font', **fontdef)
```

```python
cases = [1, 2, 3]
color_array1 = ['red', 'brown', 'seagreen', 'blue', 'purple', 'orange']
color_array2 = ['salmon','peachpuff', 'springgreen', 'aqua', 'violet', 'moccasin']
dcName = ["Constant Sample Size 1", "Constant Sample Size 3", "Constant Sample Size 5" , "Constant Sample Size 7", "Constant Sample Size 10", "Uniform Product Distribution"]
dcName2 = ["s=1", "s=3", "s=5", "s=7", "s=10", "UnifPD"]
dcNum = 6
for case in cases:
    f = open("case"+str(case)+"_out.txt", "r")
    MAX_ITER = 100
    lines = f.readlines()
    samplesize = [5, 10, 20, 40, 80, 160, 320, 640, 1280, 2560]
    opt = float(lines[1].split()[-1])
    opteq = float(lines[3].split()[-1])

    burn_direct = []
    burn_direct_error = []
    udirect_array = []
    udirect_error_array = []
    uindirect_error_array = []
    uindirect_array = []
    for dc in range(dcNum):
        util_direct = [float(item) for item in lines[20*dc + 6].split()[2:]]
        util_direct_error = [np.sqrt(float(item)/MAX_ITER) for item in lines[20*dc + 7].split()[2:]]
        util_indirect = [float(item) for item in lines[20*dc + 17].split()[2:]]
        util_indirect_error = [np.sqrt(float(item)/MAX_ITER) for item in lines[20*dc + 18].split()[2:]]
        
        udirect_array.append(util_direct)
        udirect_error_array.append(util_direct_error)
        uindirect_error_array.append(util_indirect_error)
        uindirect_error_array.append(util_indirect)
        
        burn_direct.append([float(item) for item in lines[20*dc + 14].split()[2:]])
        burn_direct_error.append([np.sqrt(float(item)/MAX_ITER) for item in lines[20*dc + 15].split()[2:]])


        util_direct = np.array(util_direct)
        util_direct_error = np.array(util_direct_error)
        util_indirect = np.array(util_indirect)
        util_indirect_error = np.array(util_indirect_error)
        
        plt.clf()
        plt.semilogx(samplesize, [opt]*len(samplesize), color = 'black', linestyle = 'dashed', label = "OPT" )
        plt.semilogx(samplesize, [opteq]*len(samplesize), color = 'grey', linestyle = 'dashed', label = "OPTEQ" )
        plt.semilogx(samplesize, util_direct, 'r-o', label = "DLE")
        plt.fill_between(samplesize, util_direct - util_direct_error, util_direct + util_direct_error, color = color_array2[0])
        plt.semilogx(samplesize, util_indirect, 'b-x', label = "ILO")
        plt.fill_between(samplesize, util_indirect - util_indirect_error, util_indirect + util_indirect_error, color = color_array2[3])
#         plt.title("Distribution: "+str(dcName[dc]))
        plt.xlabel("Number of Samples", **font)
        plt.ylabel("Total Welfare", **font)
        plt.axis(xmin = 5, ymin = 0)
        plt.legend()
        plt.savefig("Graphs/utilcase"+str(case)+"dc"+str(dc)+".pdf", bbox_inches = 'tight')

    plt.clf()
    burn_direct = np.array(burn_direct)
    burn_direct_error = np.array(burn_direct_error)
    for dc in range(dcNum):
        plt.semilogx(samplesize, burn_direct[dc, :], '-o', color = color_array1[dc], label = dcName2[dc])
        plt.fill_between(samplesize, burn_direct[dc, :] - burn_direct_error[dc, :], burn_direct[dc, :] + burn_direct_error[dc, :], color = color_array2[dc])

#     plt.title("Market: ("+chr(96+case)+")")
    plt.xlabel("Number of Samples", **font)
    plt.ylabel("Number of Goods Burnt", **font)
#     plt.legend()
    plt.savefig("Graphs/burncase"+str(case)+".pdf", bbox_inches = 'tight')
    
    plt.clf()
    burn_direct = np.array(udirect_array)
    burn_direct_error = np.array(udirect_error_array)
    for dc in range(dcNum):
        plt.semilogx(samplesize, burn_direct[dc, :], '-o', color = color_array1[dc], label = dcName2[dc])
        plt.fill_between(samplesize, burn_direct[dc, :] - burn_direct_error[dc, :], burn_direct[dc, :] + burn_direct_error[dc, :], color = color_array2[dc])

#     plt.title("Market: ("+chr(96+case)+")")
    plt.xlabel("Number of Samples", **font)
    plt.ylabel("Total Welfare", **font)
#     plt.legend()
    plt.savefig("Graphs/utilcase"+str(case)+".pdf", bbox_inches = 'tight')
    
cases = [1, 2, 3]
color_array1 = ['red', 'brown', 'seagreen', 'blue', 'purple', 'orange']
color_array2 = ['salmon','peachpuff', 'springgreen', 'aqua', 'violet', 'moccasin']
dcName = ["Constant Sample Size 1", "Constant Sample Size 3", "Constant Sample Size 5" , "Constant Sample Size 7", "Constant Sample Size 10", "Uniform Product Distribution"]
dcNum = 6
for case in cases:
    f = open("case"+str(case)+"_out.txt", "r")
    MAX_ITER = 100
    lines = f.readlines()
    samplesize = [5, 10, 20, 40, 80, 160, 320, 640, 1280, 2560]
    opt = float(lines[1].split()[-1])
    opteq = float(lines[3].split()[-1])

    burn_direct = []
    burn_direct_error = []
    udirect_array = []
    udirect_error_array = []
    uindirect_error_array = []
    uindirect_array = []
    for dc in range(dcNum):
        util_direct = [float(item) for item in lines[20*dc + 10].split()[2:]]
        util_direct_error = [np.sqrt(float(item)/MAX_ITER) for item in lines[20*dc + 11].split()[2:]]
        util_indirect = [float(item) for item in lines[20*dc + 17].split()[2:]]
        util_indirect_error = [np.sqrt(float(item)/MAX_ITER) for item in lines[20*dc + 18].split()[2:]]
        
        udirect_array.append(util_direct)
        udirect_error_array.append(util_direct_error)
        uindirect_error_array.append(util_indirect_error)
        uindirect_error_array.append(util_indirect)
        
        burn_direct.append([float(item) for item in lines[20*dc + 14].split()[2:]])
        burn_direct_error.append([np.sqrt(float(item)/MAX_ITER) for item in lines[20*dc + 15].split()[2:]])


        util_direct = np.array(util_direct)
        util_direct_error = np.array(util_direct_error)
        util_indirect = np.array(util_indirect)
        util_indirect_error = np.array(util_indirect_error)
        
        
    
    plt.clf()
    burn_direct = np.array(udirect_array)
    burn_direct_error = np.array(udirect_error_array)
    for dc in range(dcNum):
        plt.semilogx(samplesize, burn_direct[dc, :], '-o', color = color_array1[dc], label = dcName2[dc])
        plt.fill_between(samplesize, burn_direct[dc, :] - burn_direct_error[dc, :], burn_direct[dc, :] + burn_direct_error[dc, :], color = color_array2[dc])

#     plt.title("Market: ("+chr(96+case)+")")
    plt.xlabel("Number of Samples", **font)
    plt.ylabel("Empirical Loss", **font)
#     plt.legend()
    plt.savefig("Graphs/losscase"+str(case)+".pdf", bbox_inches = 'tight')
   
```
